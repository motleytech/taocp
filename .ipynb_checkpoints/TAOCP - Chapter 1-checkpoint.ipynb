### Euclid's Algorithm

    E1. r = m % n
    E2. if r == 0: return n
    E3. m, n = n, r

<hr>
Working through the example for m = 119, n = 544

    E1. r = 119
    E2. r != 0                 # step skipped
    E3. m = 544, n = 119       # if n > m: m and n exchanged in first pass

    E1. r = 68
    E3. m = 119, n = 68

    E1. r = 51
    E3. m = 68, n = 51

    E1. r = 17
    E3. m = 51, n = 17

    E1. r = 0
    E2. n = 17 = GCD

“Assuming that the value of n is known but m is allowed to range over all positive integers, what is the average number of times, Tn, that step E1 of Algorithm E will be performed?”

Lets write a program to find that...


```python
import math

def countAvgSteps(n):
    count = 0
    for m in range(0, n):
        steps = 1   
        p, q = n, m
        while q != 0:
            p = p % q
            p, q = q, p
            steps += 1
        count += steps
    return float(count) / n

for n in range(50, 1001, 50):
    print('%s, %6.4f' % (n, countAvgSteps(n))) #, 1.2783 + 0.72856*math.log(n)))
    
```
